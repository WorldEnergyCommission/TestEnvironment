```python
# configurations
y1y_range = [25, 130]
y2y_range = [20, 65]
x_range = [-15, 15] #locations of y1 and y2 on y axis
diff = -5 # min diff between y1y and Y2y

# input
x=-10 # current outside temp
y=50 # VLT soll
epsilon = 0.2

def get_y1_y2(x,y, y1y_range, y2y_range, x_range, diff):
    print(x, y)
    y1x = x_range[0]
    y2x = x_range[1]

    def check_range(val, range):
            if val > range[1]:
                return range[1]
            elif val < range[0]:
                return range[0]
            else:
                return val
            
    # Equation
    def equation(y1y, y2y):
        return ((y2y - y1y) / (y2x - y1x)) * (x-x_range[0]) + y1y
    # Constraints
    def constraints(y1y, y2y):
        return (y2y <= y1y + diff) and \
            (y1y_range[0] <= y1y <= y1y_range[1]) and \
                (y2y_range[0] <= y2y <= y2y_range[1])
    
    res = [(y1y,y2y,y2y-y1y) for y1y in range(y1y_range[0],y1y_range[1]) 
        for y2y in range(y2y_range[0],y2y_range[1]) 
        if (abs(y - equation(y1y, y2y))<epsilon) and constraints(y1y, y2y)]
    
    if not res:
        if y >= y2y_range[1]:
            y1y = y1y_range[1]
            y2y = y2y_range[1]
            print("Warning heating curve set to max")
        elif y <= y2y_range[0]:
            y1y = y1y_range[0]
            y2y = y2y_range[0]
            print("Warning heating curve set to min")
        else:
            print("unknown error")
    else:
        res_sorted = sorted(res, key=lambda x: x[2])[::-1]
        print(res_sorted)
        y1y, y2y = res_sorted[0][0:2]

    print(y1y, y2y)
    print("-----------------")
    
```
